```python
#metod naimen'shih kvadratov
def matrix_M(x, m): # m Степень полинома 
    model = []
    for i in range(m+1): 
        model.append(x**(m-i))
    model = np.stack(model).T
    return model

def mnq(x, y, m):
    model = matrix_M(x, m)
    s = np.linalg.lstsq(model, y)[0]
    return s
```
